```python
import pandas as pd
```

```python
training_data=pd.read_csv('train.csv')
```

```python
training_data.head(2)
```

```python

# Function to convert DataFrame row to Alpaca format
def convert_to_alpaca_format(row):
    instruction = "Summarize the given content."
    input_text = row['input']
    response = row['output']

    alpaca_format = f"""Below is an instruction that describes a task, paired with an input that provides further context. Write a response that appropriately completes the request.

### Instruction:
{instruction}

### Input:
{input_text}

### Response:
{response}
"""
    return alpaca_format

# Apply the function to each row and store the results in a new column
training_data['text'] = training_data.apply(convert_to_alpaca_format, axis=1)
```
